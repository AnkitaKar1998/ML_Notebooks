```python
# Q.3. Which country had the biggest difference between their summer and winter gold medal counts?
# Ans. Unites States

summer_gold = list(df1['Gold'])
winter_gold = list(df1['Gold.1'])
result = []
for (i,j) in zip(summer_gold, winter_gold):
    result.append(abs(i-j))
index = result.index(max(result))
print(df1.iloc[index])
print("\nCountry:-")
print(df1.index.values[index])
```

```python
# Q.4. Which country has the biggest difference between their summer gold medal counts and winter gold medal counts 
# relative to their total gold medal count? 
# Only include countries that have won at least 1 gold in both summer and winter.
# Ans. Bulgaria

summer = list(df1['Gold'])
winter = list(df1['Gold.1'])
total = list(df1['Gold.2'])

result={}

for (i,j,k) in zip(range(len(summer)), range(len(winter)), range(len(total))):
    if(summer[i]>0 and winter[j]>0):
        r=abs((summer[i]-winter[j])/total[k])
        result[df1.index.values[i]] = r
print("Relative values:-")
for i in result:
    print('{}: {}'.format(i, result[i]))
m = max(result, key=result.get)
print("\n\nMax: {}".format(m))
```

```python
# Q.5. Write a function that creates a Series called "Points" which is a weighted value where 
# each gold medal (Gold.2) counts for 3 points, silver medals (Silver.2) for 2 points, and bronze medals (Bronze.2) for 1 point.
# The function should return only the column (a Series object) which you created, with the country names as indices.
# (This function should return a Series named Points of length 146)
# Ans.

gold = list(df1['Gold.2'])
silver = list(df1['Silver.2'])
bronze = list(df1['Bronze.2'])

values = []

for (i,j,k) in zip(range(len(gold)), range(len(silver)), range(len(bronze))):
    r=3*gold[i] + 2*silver[j] + bronze[k]
    values.append(r)
    
series = pd.Series(values, index = list(df1.index), name="Points")
print(series)
```
